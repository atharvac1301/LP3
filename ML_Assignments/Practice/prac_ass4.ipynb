```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    return (x + 3)**2

def derivative(x):
    return 2*(x + 3)

def grad_descent(starting_x, learning_rate, num_iterations):
    x = starting_x
    x_history = [x]
    
    for i in range(num_iterations):
        grad = derivative(x)
        x = x - learning_rate * grad
        x_history.append(x)
        print(f"Iterations : {i}, Value : {x}")
    return x, x_history
```

```python
starting_x = 2  
lr = 0.1
n = 500

x_final, x_history = grad_descent(starting_x, lr, n)
print(f"\nLocal Minimum occurs at : {round(x_final, 4)}")
print(f"f(x) at local minima : {round(function(x_final), 4)}")
```

```python
plt.plot(x_history, function(np.array(x_history)), marker='o')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
