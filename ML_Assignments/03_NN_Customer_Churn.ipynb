```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from keras.models import Sequential
from keras.layers import Dense
```

### Optree Module Issue on Jupyter Notebook, use Colab instead

```python
df = pd.read_csv('/content/Churn_Modelling.csv')
df
```

```python
# Drop useless columns
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.Geography.unique()
```

```python
geography_dict = {'France' : 1, 'Spain' : 2, 'Germany' : 3}
gender_dict = {'Female' : 0, 'Male' : 1}

for i in range(10000):
    df['Geography'][i] = geography_dict[df['Geography'][i]]
    df['Gender'][i] = gender_dict[df['Gender'][i]]

df['Geography'] = df['Geography'].astype(int)
df['Gender'] = df['Gender'].astype(int)
```

```python
y = df['Exited']
X = df.drop(['Exited'], axis=1)
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
model = Sequential(
    [
        Dense(8, activation='relu'),
        Dense(11, activation='relu'),
        Dense(6, activation='relu'),
        Dense(3, activation='relu'),
        Dense(1, activation='sigmoid')
    ]
)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, batch_size=10, epochs=50)
```

```python
threshold = 0.5
y_pred = model.predict(X_test)
y_pred = (y_pred > threshold)
```

```python
accuracy_score(y_test, y_pred)
```
