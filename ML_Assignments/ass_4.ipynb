```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def function(x):
    return (x + 3)**2

def derivative(x):
    return 2 * (x + 3)

def gradient_descent(starting_x, learning_rate, num_iterations):
    x = starting_x
    x_history = [x]
    
    for i in range(num_iterations):
        grad = derivative(x)
        x = x - learning_rate * grad
        x_history.append(x)
        print(f"Iterations: {i}, Value: {x}")
    
    return x, x_history

# Set parameters
starting_x = 2
learning_rate = 0.1
num_iterations = 500

# Run gradient descent
final_x, x_history = gradient_descent(starting_x, learning_rate, num_iterations)

# Print results
print(f"Local minimum occurs at x = {final_x:.4f}")
print(f"f(x) at local minimum = {function(final_x):.4f}")
```

```python
# Plotting
x = np.linspace(-6, 6, 100)
y = function(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'r-', label='f(x) = (x+3)^2')
plt.scatter(x_history, [function(x) for x in x_history], c='blue', label='Gradient descent path')
plt.title('Gradient Descent Visualization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
